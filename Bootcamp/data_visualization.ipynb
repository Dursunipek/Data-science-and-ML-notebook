```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# Matplotlib

```python
plt.plot(range(14),[i**2 for i in range(14)],"r*")
plt.xlabel("numbers")
plt.ylabel("square of numbers")
plt.title("plt.plot")
plt.show()
```

```python
# Plot fonksiyonuna tek figure üzerine birden fazla grafik cizdirebiliriz
arr = np.arange(0,10,0.5)
plt.plot(arr,arr,"r*",arr,arr**2,"b*",arr,arr**3,"g*",markersize=7)
#markersize -> işaretlerin büyüklüğü
#linewidth -> çizgi kalınlığı
plt.show()
```

```python
data = {"a": np.arange(50),
        "b": np.random.randint(0,100,50),
        "c": np.random.randn(50)}
data["d"] = data["a"] + (10 * np.abs(np.random.randn(50)))
data["c"] = np.abs(data["c"]) * 100

plt.scatter(data=data,x="a",y="b",c="c",s="d")
plt.show()
```

```python
# data
players = ["player1","player2","player3","player4"]
points = np.random.randint(0,100,4)

# subplot -> Birden fazla grafiği bir arada cizmek icin kullanilir
plt.figure(figsize=(9,3))

# graph 1
plt.subplot(131) # 131 -> 1 satir 3 sütun olan figürün 1. grafiği
plt.bar(players,points,color="orange")

# graph2
plt.subplot(132) # 132 -> 1 satır 3 sütun olan figürün 2. graifiği
plt.scatter(players,points)

#graph3
plt.subplot(133) # 133 -> 1 satir 3 sütun olan figürün 3. grafiği
plt.plot(players,points,"r--")

plt.show()
```

```python
#histogram
mu1 = 100 # ortalama
sigma1 = 15 # standart sapma
x = mu1 + (sigma1 * np.random.randn(10000)) # normal dağılım olan bir veri kümesi

mu2 = 110 
sigma2 = 10
y = mu2 + (sigma2 * np.random.randn(10000))

plt.figure(figsize=(9,5))
plt.hist(x,bins=100,density=1,facecolor="g",alpha=0.25) # bins -> histogramın aralık sayisi
plt.hist(y,bins=100,density=1,facecolor="b",alpha=0.25)
plt.xlabel("Data")
plt.ylabel("Probabilities")
plt.title(r"$\mu_1=100,\ \mu_2=110$") # Mü, sigma gibi işaretleri 2 $ işareti arasinda yazdirabiliyoruz
plt.grid(True,alpha=0.4) # Grid oluşturuldu
plt.annotate('Top of the gaussian curve',xy=(110,0.04),xytext=(130,0.045),
            arrowprops=dict(facecolor='black',shrink=1)) # Ok işareti ile histogramın tepe noktası işaret edildi
plt.text(60,0.020,r"$\mu_1=100,\ \sigma_1=15$") 
plt.text(120,0.030,r"$\mu_1=110,\ \sigma_2=10$")
plt.axis([40,160,0,0.05])
plt.show()
```

```python
# Scaling
# Bazı durumlarda sayilar cok büyük olduğu için grafik uzerinde anlasılamaz olabilir
# Bu sorunu gidermek icin logaritma fonksiyonu ile y eksenindeki degerler scale edilerek algılanabilir hale getirildi
x = np.arange(0,10,0.01)
y = np.exp(x)

plt.figure(figsize=(9,3))

plt.subplot(121)
plt.plot(x,y)
plt.yscale("linear")
plt.title("Linear scaling")
plt.grid(True,alpha=0.4)

plt.subplot(122)
plt.plot(x,y)
plt.yscale("log")
plt.title("Logaritmic scaling")
plt.grid(True,alpha=0.4)

plt.show()
```

# Seaborn

```python
# import data
df = pd.read_csv("../datasets/clean_athlete_events.csv")
df.head()
```

```python
df.info()
```

## Scatter Plot

```python
def drawScatterPlot():
    sns.scatterplot(x="height",y="weight",data=df)
    plt.title("Relationship between height and weight")
    sns.set_style(style="white")
    plt.show()
drawScatterPlot()
```

```python
sns.set_style(style="darkgrid")
drawScatterPlot()
```

```python
sns.set_style(style="whitegrid")
drawScatterPlot()
```

## Scatter Plot - Categorical

```python
# Kategorik veriler hue parametresi ile istenilen sütundaki farklı degerler renklendirilerek birbirinden ayrıldı
sns.scatterplot(x="height",y="weight",data=df,hue="medal")
plt.title("Height and weight scatter plot by medals")
plt.grid(True,alpha=0.2)
plt.show()
```

```python
sns.scatterplot(x="height",y="weight",data=df,hue="gender")
plt.title("Height and weight scatter plot by gender")
plt.grid(True,alpha=0.4)
plt.show()
```

```python
# Style parametresi ile grafiğe 4. boyutu kazandırıyoruz
# Style parametresi ile farklı madalya alan sporcuları farklı sekillerde gosteriyoruz
sns.scatterplot(x="height",y="weight",data=df,hue="gender",style="medal")
plt.title("Height and weight scatter plot by gender")
plt.grid(True,alpha=0.4)
plt.show()
```

```python
# Size parametresi ile 5. boyutlu grafik olusturuyoruz
sns.scatterplot(x="height",y="weight",data=df,hue="gender",style="medal",size="age")
plt.title("Height and weight scatter plot by gender")
plt.grid(True,alpha=0.4)
plt.show()
```

 ## Scatter Plot - Linear Regression   

```python
sns.regplot(x="height",y="weight",data=df,marker="+",scatter_kws={'alpha':0.2})
plt.grid(True,alpha=0.4)
plt.show()
```
