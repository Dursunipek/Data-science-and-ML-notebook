```python
newDf = df.sort_values("price",ascending=False).iloc[131:] 
```

```python
newDf = newDf[newDf.year != 1970]
```

```python
newDf.head()
```

```python
newDf = newDf.drop("model",axis=1)
```

```python
newDf = newDf.drop("transmission",axis=1)
```

```python
newDf = newDf.drop("fuelType",axis=1)
```

```python
newDf.head()
```

```python
y = newDf["price"].values
```

```python
x = newDf.drop("price",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),epochs=300,batch_size=250)
```

```python
lossData = pd.DataFrame(model.history.history)
```

```python
lossData.head()
```

```python
lossData.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
predictArray = model.predict(x_test)
```

```python
predictArray
```

```python
mean_absolute_error(y_test,predictArray)
```

```python
plt.scatter(predictArray,y_test)
plt.plot(y_test,y_test,"g",)
```
